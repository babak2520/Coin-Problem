```python
import random
import numpy as np
```

```python
#Function to calculate the amount of money made for a randomly generated set of coins from
#1 to N. Every time this function gets called, it randonmly generates numbers from 1 to N
#and spits out the total money made for that particular combination.
def MakeMoney(N):
    MyList = list(range(1, N+1))
    i = N - 1
    MoneyMade = 0
    Last = 0
    while i >= 0:
        Current = MyList.pop(random.randint(0,i))
        MoneyMade += abs(Last - Current)
        Last = Current
        i -= 1
    return MoneyMade
        
```

```python
#loop over randomly generated combinations and the amount of total money made. Some key statitstics
#get printed every XXX iterations of the loop so that convergence status can be seen.
MoneyList = []
Counter = 0
while True:
    MoneyList.append(MakeMoney(20))
    Counter += 1
    if Counter == 50000:
        print("Median money made: {}".format(np.mean(MoneyList)))
        print("Standard deviation: {}".format(np.std(MoneyList)))
        GoodMoney = 0
        for item in MoneyList:
            if item >= 160:
                GoodMoney += 1
        Possibility = GoodMoney / len(MoneyList) * 100
        print("Possibility of more than $160: {}%".format(Possibility))
        Counter = 0
        
```
